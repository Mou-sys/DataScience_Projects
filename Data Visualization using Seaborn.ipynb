```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
# load the tips dataset
data_tips=sns.load_dataset('tips')
```

```python
# remove warning in python
import warnings
warnings.filterwarnings('ignore')
```

```python
data_tips.info()
```

```python
data_tips.columns
```

```python
data_tips.head()
```

```python
# Scatter plot

# find tip according to total bill in a scatter plot

plt.figure(figsize=(7,5))
sns.scatterplot(x='total_bill',y='tip',data=data_tips)
plt.show()
```

```python
# find gender who makes the tip according to the of total bill in a scatter plot
plt.figure(figsize=(7,4))
sns.scatterplot(x='total_bill',y='tip',data=data_tips,hue='sex')
```

```python
# find per day tips according to total bill in scatter plot & save it
plt.figure(figsize=(7,5))
sns.scatterplot(x='total_bill',y='tip',data=data_tips,hue='day')
plt.savefig('per day tips.jpg')
```

```python
# find tips of smokers according to total bill
plt.figure(figsize=(7,4))
sns.scatterplot(x='total_bill',y='tip',data=data_tips,hue='smoker')
plt.show()
```

```python
# Bar plot
# find which sex has more total bill
sns.barplot(x=data_tips['sex'],y=data_tips['tip'],ci=False)
```

```python
data_tips['total_bill'].mean()
```

```python
# find day wise maximum total bill of each sex
plt.figure(figsize=(8,5))
sns.barplot(x=data_tips['sex'],y=data_tips['total_bill'],estimator=np.max,palette='Greens',
           hue=data_tips['day'],ci=None)
plt.show()
```

```python
data_tips['total_bill'].max()
```

```python
# countplot
# find which time mostly people are coming into the hotel each day
data_tips['day'].value_counts()
```

```python
sns.countplot(x=data_tips['day'],hue=data_tips['time'])
```

```python
# find which sex is coming more on which day
sns.countplot(x=data_tips['day'],hue=data_tips['sex'])
```

```python
# Strip plot
# how much each sex spends on each day
sns.stripplot(x=data_tips['sex'],y=data_tips['total_bill'],hue=data_tips['day'])
```

```python
# Swarm plot
# how much each sex spends on each day
sns.swarmplot(x=data_tips['sex'],y=data_tips['total_bill'],hue=data_tips['day'])
```

```python
# Pair plot
# See all the graphs of numerical data
sns.pairplot(data_tips)
```

```python
sns.pairplot(data_tips,hue='sex')
```

```python
data_tips.corr()
```

```python
# Heat map
# show correlation b/w numerical data
sns.heatmap(data_tips.corr(),annot=True)   # annot prints the values on chart
```

```python
sns.heatmap(data_tips.corr(),annot=True,cmap='magma') # cmap changes color palette
```

```python
# Box plot
# finf outliers in total_bill
sns.boxplot(data_tips['total_bill'])
```

```python
# which time gets higher total bill
sns.boxplot(x=data_tips['time'],y=data_tips['total_bill'],data=data_tips)
```

```python
# which time gets higher total bill from each sex
plt.figure(figsize=(8,5))
sns.boxplot(x=data_tips['time'],y=data_tips['total_bill'],hue=data_tips['sex'])
```

```python
# Violin plot
# which time gets higher total bill from each sex
plt.figure(figsize=(7,4))
sns.violinplot(x=data_tips['time'],y=data_tips['total_bill'],hue=data_tips['sex'])
```

```python
# Distribution plot
# plot distribution of total_bill
sns.distplot(data_tips['total_bill'])
```
